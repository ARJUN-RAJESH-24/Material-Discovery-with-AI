# Baseline ML Workflow

This notebook demonstrates the classical ML baseline using composition features (Magpie) and a gradient boosted model.

## Steps
1. Load dataset from `data/`
2. Ensure `composition` and target columns exist
3. Train baseline (XGBoost/LightGBM/RF)
4. Save model and metrics to `results/`

## CLI Equivalent

```bash
python scripts/baseline_ml.py \
  --dataset data/materials.csv \
  --target formation_energy_per_atom \
  --model xgboost \
  --output results/baseline_formation_energy
```
